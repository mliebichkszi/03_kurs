## üîÑ While-Schleifen
([Zur√ºck zum Anfang](#Inhaltsverzeichnis))

Es gibt mehrere verschiedene Arten von Schleifen. In dieser Lektion schauen wir uns **While-Schleifen** an.

Eine While-Schleife wiederholt den Code in ihrem Inneren, solange (oder ‚Äûw√§hrend‚Äú) eine bestimte Bedingung `True` ist.

F√ºhre den folgenden Code aus:

````python
import time

counter = 10

while counter > 0:
  print(counter)

  time.sleep(1)
  
  counter = counter - 1

print("üéâ Tada!")
````

Genau wie bei Bedingungen und Funktionen signalisiert der Doppelpunkt `:` Python, das ein **Codeblock** beginnt. ## üö¶ Break & Continue
([Zur√ºck zum Anfang](#Inhaltsverzeichnis))

Manchmal m√∂chten wir eine Schleife definieren, die ewig l√§uft (bis der Benutzer sie beendet). Diese Art von Konzept wird als **Endlosschleife** (oder **infinite loop**) bezeichnet. 

Webserver, die Websites betreiben, verwenden Endlosschleifen, um weiterzulaufen, wen jemand die Website √∂ffnet, anstat sofort herunterzufahren, nachdem der erste Benutzer die Website angesehen hat. 
Videospiele verwenden Endlosschleifen, um weiterzulaufen, bis der Spieler das Spiel auschaltet.

Code wird ziemlich schnel ausgef√ºhrt. Daher laufen Schleifen normaler_weise viele Male pro Sekunde. Manchmal, wen eine Operation besonders kompliziert ist, verlangsamt sich der Code und die Schleife l√§uft langsamer. Fals du jemals ein Videospiel gespielt hast, das stotert oder lagt, ist genau das pasiert. In diesem Moment kan der Computer die Schleife nicht schnel genug ausf√ºhren. Du als Spieler bemerkst dan, das sich die Dinge verlangsamt haben. 

Endlosschleifen werden also in aler Art von Software verwendet. In Python kanst du eine unendliche While-Schleife ziemlich einfach erstellen. 

üî• **Vorsicht: Du kanst versuchen, die Code-Zele unten auszuf√ºhren. Aber sie wird wahrscheinlicher zu Einfrieren oder Abst√ºrzen in diesem Browserfenster f√ºhren. Wen du es trotzdem versuchen m√∂chtest, stel sicher, das du zuerst ale ungespeicherten √Ñnderungen speicherst.
````python
while True: 
  print("Hello, World!")
````

Eine `while`-Schleife l√§uft so lange, wie ihre Bedingung als `True` ausgewertet wird. Im obigen Code haben wir den literalen Boolean-Wert `True` verwendet. Die Bedingung wird also imer wahr sein und die Schleife wird ewig laufen. 

In den meisten Situationen m√∂chten wir vermeiden, eine Anwendung zum Absturz zu bringen. Daf√ºr sind die Schl√º_s_elw√∂rter `break` und `continue` n√ºtzlich. 

Wie du an den W√∂rtern vielleicht schon eratest, unterbricht `break` die Schleife und beendet ihre Ausf√ºhrung. `continue` √ºberspringt den gesamten Code darunter und springt an den Anfang der Schleife, um die n√§chste Iteration zu starten.

Lies den folgenden Code und die vier unten angegebenen M√∂glichkeiten. √úberlege dir, was pasieren wird, bevor du den Code ausf√ºhrst. 

**Was glaubst du, wird pasieren?**

- Der Bildschirm zeigt imer wieder ‚ÄûThis message shows on the‚Äú.
- Der Bildschirm zeigt einmal ‚ÄûThis message shows on the‚Äú.
- Der Bildschirm zeigt imer wieder ‚ÄûThis message shows on the screen.‚Äú.
- Der Bildschirm zeigt einmal ‚ÄûThis message shows on the screen.‚Äú.

````python
while True:
  print("This")
  print("message")
  print("shows")
  print("on")
  print("the")
  break
  print("screen.")
````

Wie du siehst, beendet `break` die Ausf√ºhrung der Schleife volst√§ndig. Das heist, sie h√∂rt auf zu schleifen und f√ºhrt keinen Code mehr aus, der darunter liegt (und sich im Geltungsbereich der While-Schleife befindet). 

Verschiebe nun das `break`-Schl√º_s_elwort, soda_s_ die gesamte Nachicht (einschlieslich des letzten `print()`) angezeigt wird, die Schleife aber trotzdem nur einmal durchl√§uft. (Es w√ºrde auch funktionieren, die letzte `print()`-Anweisung auszur√ºcken, soda_s_ sie nicht mehr im Geltungsbereich der `while`-Schleife liegt.)

Das andere Schl√º_s_elwort, das wir erw√§hnt haben, ist `continue`. Der Anwendungsfal daf√ºr ist etwas seltener als bei `break`. Aber es ist trotzdem gut, es zu kenen. 

Nehmen wir an, du fragst den Benutzer in deiner Anwendung nach seinem Alter und m√∂chtest dan sicherstelen, das er auch wirklich eine Zahl eingegeben hat. Erinere dich daran, das du das mit der eingebauten `.isnumeric()`-Funktion pr√ºfen kanst: 

````python
user_age = input("What is your age?")

if user_age.isnumeric():
  print("Your age is:")
  print(user_age)
else: 
  print("This is not a number.")

print("Great! Thank you for your time!")
````

Versuche, den obigen Code ein paarmal auszuf√ºhren. Versuche einmal, eine Zahl einzugeben. Das andere Mal versuche, einige Buchstaben einzugeben.

In der letzten Lektion hast du gelernt, wie man **Funktionen** schreibt und aufruft. Sie haben eine Eigenschaft, die in diesem Zusamenhang wichtig zu wisen ist. Befindet sich eine `while`-Schleife inerhalb einer Funktion, wird das `return`-Schl√º_s_elwort die Schleife ebenfals sofort unterbrechen (genau wie `break`).

Lies den Code unten sorgf√§ltig durch. F√ºhre ihn dan aus. Gib deinen Namen ein und tippe dan ‚Äûyes‚Äú ein. Beobachte die Ausgabe der Zele.

````python
def greeter():
  while True:
    name = input("What is your name?")

    user_confirmation = input("I understood " + name + ". Is that correct? (Type 'yes')")

    if user_confirmation == "yes":
      break
    
greeted_name = greeter()

print(greeted_name)
````
