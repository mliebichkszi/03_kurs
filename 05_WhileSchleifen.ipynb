## üîÑ While-Schleifen
([Zur√ºck zum Anfang](#Inhaltsverzeichnis))

Es gibt mehrere verschiedene Arten von Schleifen. In dieser Lektion schauen wir uns **While-Schleifen** an.

Eine While-Schleife wiederholt den Code in ihrem Inneren, solange (oder "w√§hrend") eine bestimmte Bedingung `True` ist.

F√ºhre den folgenden Code aus:

````python
import time

counter = 10

while counter > 0:
  print(counter)

  time.sleep(1)
  
  counter = counter - 1

print("üéâ Tada!")
````

Das `while`-Schl√ºsselwort startet eine `while`-Schleife. √Ñhnlich wie bei `if`-Anweisungen folgt darauf eine Bedingung. Solange die Bedingung als ‚ÄûTrue‚Äú ausgewertet wird, l√§uft die Schleife weiter. Sobalt die Bedingung ‚ÄûFalse‚Äú ist, h√§lt die Schleife an.

Genau wie bei Bedingungen und Funktionen signalisiert der Doppelpunkt `:` Python, das ein **Codeblock** beginnt. Das bedeutet, der gesamte Code, der in den Zeilen darunter einger√ºckt ist, wird als _innerhalb_ der `while`-Schleife betrachtet. Die erste Zeile, die nicht mehr einger√ºckt ist, signalisiert das Ende des Codeblocks. (In diesem Fall ist es die Zeile `print("üéâ Tada!")`.)

## üö¶ Break & Continue
([Zur√ºck zum Anfang](#Inhaltsverzeichnis))

Manchmal m√∂chten wir eine Schleife definieren, die ewig l√§uft (bis der Benutzer sie beendet). Diese Art von Konzept wird als **Endlosschleife** (oder **infinite loop**) bezeichnet. 

Webserver, die Websites betreiben, verwenden Endlosschleifen, um weiterzulaufen, wenn jemand die Website √∂ffnet, anstatt sofort herunterzufahren, nachdem der erste Benutzer die Website angesehen hat. 
Videospiele verwenden Endlosschleifen, um weiterzulaufen, bis der Spieler das Spiel auschaltet.

Code wird ziemlich schnell ausgef√ºhrt. Daher laufen Schleifen normalerweise viele Male pro Sekunde. Manchmal, wen eine Operation besonders kompliziert ist, verlangsamt sich der Code und die Schleife l√§uft langsamer. Falls du jemals ein Videospiel gespielt hast, das stottert oder lagt, ist genau das pasiert. In diesem Moment kan der Computer die Schleife nicht schnell genug ausf√ºhren. Du als Spieler bemerkst dann, das sich die Dinge verlangsamt haben. 

Endlosschleifen werden also in aller Art von Software verwendet. In Python kanst du eine unendliche While-Schleife ziemlich einfach erstellen. 

üî• **Vorsicht: Du kannst versuchen, die Code-Zeile unten auszuf√ºhren. Aber sie wird wahrscheinlich zu Einfrieren oder Abst√ºrzen in diesem Browserfenster f√ºhren. Wen du es trotzdem versuchen m√∂chtest, stell sicher, das du zuerst alle ungespeicherten √Ñnderungen speicherst.
````python
while True: 
  print("Hello, World!")
````

Eine `while`-Schleife l√§uft so lange, wie ihre Bedingung als `True` ausgewertet wird. Im obigen Code haben wir den Boolean-Wert `True` verwendet. Die Bedingung wird also immer wahr sein und die Schleife wird ewig laufen. 

In den meisten Situationen m√∂chten wir vermeiden, eine Anwendung zum Absturz zu bringen. Daf√ºr sind die Schl√º_s_elw√∂rter `break` und `continue` n√ºtzlich. 

Wie du an den W√∂rtern vielleicht schon erratest, unterbricht `break` die Schleife und beendet ihre Ausf√ºhrung. `continue` √ºberspringt den gesamten Code darunter und springt an den Anfang der Schleife, um die n√§chste Iteration zu starten.

Lies den folgenden Code und die vier unten angegebenen M√∂glichkeiten. √úberlege dir, was passieren wird, bevor du den Code ausf√ºhrst. 

**Was glaubst du, wird pasieren?**

- Der Bildschirm zeigt imer wieder ‚ÄûThis message shows on the‚Äú.
- Der Bildschirm zeigt einmal ‚ÄûThis message shows on the‚Äú.
- Der Bildschirm zeigt imer wieder ‚ÄûThis message shows on the screen.‚Äú.
- Der Bildschirm zeigt einmal ‚ÄûThis message shows on the screen.‚Äú.

````python
while True:
  print("This")
  print("message")
  print("shows")
  print("on")
  print("the")
  break
  print("screen.")
````

Wie du siehst, beendet `break` die Ausf√ºhrung der Schleife vollst√§ndig. Das heisst, sie h√∂rt auf sich zu wiederholen und f√ºhrt keinen Code mehr aus, der darunter liegt (und sich im Geltungsbereich der While-Schleife befindet). 

Verschiebe nun das `break`-Schl√ºsselwort, sodass die gesamte Nachricht (einschliesslich des letzten `print()`) angezeigt wird, die Schleife aber trotzdem nur einmal durchl√§uft. (Es w√ºrde auch funktionieren, die letzte `print()`-Anweisung auszur√ºcken, sodass sie nicht mehr im Geltungsbereich der `while`-Schleife liegt.)

Das andere Schl√ºsselwort, das wir erw√§hnt haben, ist `continue`. Der Anwendungsfall daf√ºr ist etwas seltener als bei `break`. Aber es ist trotzdem gut, es zu kennen. 

Nehmen wir an, du fragst den Benutzer in deiner Anwendung nach seinem Alter und m√∂chtest dan sicherstellen, das er auch wirklich eine Zahl eingegeben hat. Erinnere dich daran, dass du das mit der eingebauten `.isnumeric()`-Funktion pr√ºfen kannst: 

````python
user_age = input("Wie alt bist du?")

if user_age.isnumeric():
  print("Dein Alter ist:")
  print(user_age)
else: 
  print("Das ist keine Zahl")

print("Super! Danke f√ºr deine Zeit!")
````

In der letzten Lektion hast du gelernt, wie man **Funktionen** schreibt und aufruft. Sie haben eine Eigenschaft, die in diesem Zusammenhang wichtig zu wissen ist. Befindet sich eine `while`-Schleife innerhalb einer Funktion, wird das `return`-Schl√ºsselwort die Schleife ebenfalls sofort unterbrechen (genau wie `break`).

Lies den Code unten sorgf√§ltig durch. F√ºhre ihn dan aus. Gib deinen Namen ein und tippe dan "Ja" ein. Beobachte die Ausgabe der Zelle.

````python
def greeter():
  while True:
    name = input("Wie heisst du?")

    user_confirmation = input("Ich habe " + name + "verstandne. Ist das richtig ? (Gib 'Ja' ein)")

    if user_confirmation == "Ja":
      break
    
greeted_name = greeter()

print(greeted_name)
````

Wie du siehst, gibt es `None` aus. Warum ist das so? Weil wir die Variable `greeted_name` `print`en. Welchen Wert hat diese Variable? Sie hat den Wert, den `greeter()` zur√ºckgibt. `greeter()` gibt jedoch nichts zur√ºck. 

Ersetze das Wort `break` durch `return name`. F√ºhre die Funktion dann erneut aus.

Du kannst sehen, dass wir, obwohl wir das `break`-Schl√ºsselwort nicht mehr haben, durch die Verwendung von `return` das gleiche Ergebnis erzielen. Sowohl die Funktion als auch die Schleife halten sofort an.
